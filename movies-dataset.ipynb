```python
# import dependencies
import pandas as pd
import numpy as np
import ast
```

```python
# load csv files
credits_df = pd.read_csv("Resources/credits.csv")
keywords_df = pd.read_csv("Resources/keywords.csv")
metadata_df = pd.read_csv("Resources/movies_metadata.csv")
```

```python
# clean data - we need id, title, cast, crew, genres, keywords, and overview
movies = metadata_df[["title", "genres", "overview"]]
initial_merge = pd.merge(keywords_df, credits_df, on="id")
initial_merge.head()
```

```python
second_merge = pd.concat([initial_merge, movies], axis=1)
second_merge.head()
```

```python
# rearrange columns for visibility
mini_meta = second_merge[["id", "title", "genres", "cast", "crew", "keywords", "overview"]]
mini_meta.head()
```

```python
# function to extract director's name
def get_director(x):
    try: 
        for item in ast.literal_eval(x):
            if item["job"] == "Director":
                return item["name"]
            
    except:
        return np.nan
```

```python
# function to return elements in column (e.g. actor's name)
def get_elements(x):
    try:
        names = []
        for item in ast.literal_eval(x):
            names.append(item["name"])
            
    except:
        names.append("")
    
    return names
```

```python
# apply both functions to mini_meta
# (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.apply.html)
mini_meta["director"] = mini_meta['crew'].apply(get_director)
mini_meta["director"].head()
```

```python
columns = ["genres", "cast", "keywords"]
for column in columns:
    mini_meta[column] = mini_meta[column].apply(get_elements)
```

```python
mini_meta = mini_meta.drop(["crew"], axis=1)
mini_meta.to_csv("Resources/mini_meta.csv", index=False)
```

```python
mini_meta.head()
```

```python
# prep data for soup
def clean_data(x):
    if isinstance(x, list):
        return (str.lower(i.replace(" ", "")) for i in x)
    else:
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ""
```

```python
# create copy of mini_meta for soup
soup_df = mini_meta

soup_col = ["genres", "cast", "keywords", "director"]
for sc in soup_col:
    soup_df[sc] = soup_df[sc].apply(clean_data)
```

```python
# create soup
def create_soup(soup_col):
    return ' '.join(soup_col["keywords"]) + ' ' + ' '.join(soup_col["cast"]) + ' ' + ' '.join(soup_col["director"]) + ' ' + ' '.join(soup_col["genres"])

soup_df["soup"] = soup_df.apply(create_soup, axis=1)
soup_df["soup"].head()
```

```python
soup_df.head()
```

```python
soup_df.to_csv("Resources/mini_meta_soup.csv", index=False)
```

```python
# process and convert soup into vectorizer, then calculate similarity between movies to make recommendations
# don't take stopwords (e.g. 'a', 'an', 'the') into account

# (https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html)
# (https://scikit-learn.org/stable/modules/generated/sklearn.metrics.pairwise.cosine_similarity.html)
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

count_vector = CountVectorizer(stop_words="english")
count_matrix = count_vector.fit_transform(soup_df["soup"])

print(count_matrix.shape)
```

```python
# (https://www.datasciencelearner.com/sklearn-cosine-similarity-implementation/)
cosine_score = cosine_similarity(count_matrix, count_matrix)
print(cosine_score.shape)
```

```python
# grab title and index (helps with grabbing user input and making recommendations)
soup_df = soup_df.reset_index()

index_df = pd.Series(soup_df.index, index=soup_df["title"]).drop_duplicates()
index_df.head()
```

```python
# build function for movie recommendations 
# input: movie title and cosine similarity score
# output: 5 recommendations
def get_recommendations(title, cosine_score=cosine_score):
    # retrieve movie index using title
    idx = index_df[title]
    # list of similarity scores - tuples contain index and cosine similarity
    scores_list = list(enumerate(cosine_score[idx]))
    # sort list in descending order by similarity score
    scores_list = sorted(scores_list, key=lambda x: x[1], reverse=True)
    # grab similarity score for top 5 recommendations
    scores_list = scores_list[1:6]
    # grab index for top 5 recommendations
    movie_index = [i[0] for i in scores_list]
    # map index to respective title and return recommendations
    movie_recs = soup_df["title"].iloc[movie_index]
    return movie_recs
```
